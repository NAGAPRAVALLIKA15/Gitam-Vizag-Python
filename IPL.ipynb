# Mini project on IPL dataset

```python
import pandas as pd
filePath='data files/Match.csv'
def readCsvData(filePath):
    return pd.read_csv(filePath)
df=readCsvData(filePath)
print(df)
```

### 1.Display the count of matches based on the year wise

```python
#Display the count of matches based on the year wise
df. groupby('Season_Year')['Match_SK'].count()
```

### Plot a bar graph  no. of matches won by each team in the year  2011

```python
C=df.groupby('Season_Year')
B=C.get_group(2011)
c=B.groupby('match_winner')['Season_Year'].count()
print(c)
D=c.tolist()
print(D)
```

```python
import numpy as np
import matplotlib.pyplot as plt
marks=D
divisions=['Chennai Super Kings','Deccan Chargers', 'Delhi Daredevils','Kings XI Punjab','Kochi Tuskers Kerala','Kolkata Knight Riders','Mumbai Indians'                 
,'Pune Warriors'                   
,'Rajasthan Royals'                
,'Royal Challengers Bangalore']
index=np.arange(len(divisions))
plt.barh(divisions,marks,color="#1412cd")
plt.show()
```

### type 2

```python
pt=df[df['Season_Year']==2011]
lr=pt['match_winner'].value_counts()
pd.DataFrame(lr)
print(lr)
print()
teams=[]
teams=lr.index.tolist()
plt.barh(teams,lr)
```
